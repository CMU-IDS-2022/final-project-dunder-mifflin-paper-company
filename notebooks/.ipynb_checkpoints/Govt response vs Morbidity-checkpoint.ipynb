### correlation

```python
location = "US_CA"
stringency_factors = ["stay_at_home_requirements", "facial_coverings", "restrictions_on_gatherings"]

rolling_window = 7
start_date = datetime.date(2021,1,1)
end_date = datetime.date(2021,12,31)
```

```python
location_cases = cases_df[cases_df['location_key'] == location][["date", "new_confirmed"]]
location_cases = location_cases.set_index(['date'])
location_cases = location_cases.loc[start_date : end_date]
cases_values = location_cases["new_confirmed"].rolling(rolling_window).mean()[rolling_window - 1:]
plt.plot(range(len(cases_values)), cases_values)
```

```python
location_government_response = government_response_df[government_response_df['location_key'] == location]
location_government_response = location_government_response.set_index(['date'])
location_government_response = location_government_response.loc[start_date : end_date]
stringency_values = location_government_response[stringency_factors][rolling_window - 1:].sum(axis = 1)
plt.plot(range(len(stringency_values)), stringency_values)
```

```python
corref = np.corrcoef(
    x = stringency_values,
    y = cases_values
)

print ("Lag:", 0, "Correlation:", corref[0, 1])
    
for shift in range(1, 120):
    corref = np.corrcoef(
        x = stringency_values[:-shift],
        y = cases_values[shift:]
    )
    
    print ("Lag:", shift, "Correlation:", corref[0, 1])
```

```python
shift = 45

shifted_stringency_values = stringency_values[:-shift]
shifted_stringency_values /= max(shifted_stringency_values)

shifted_case_values = cases_values[shift:]
shifted_case_values /= max(shifted_case_values)

corref = np.corrcoef(
    x = shifted_stringency_values,
    y = shifted_case_values
)

print ("Correlation coeff:", corref[0, 1])

plt.plot(range(len(shifted_case_values)), shifted_case_values, label = "Cases with lag = " + str(shift) + " days")
plt.plot(range(len(shifted_stringency_values)), shifted_stringency_values, label = "Stringency value")
plt.legend()
```
