```python
df = pd.read_csv("../data/COVID-19_Public_Therapeutic_Locator.csv")
```

```python
df = df.dropna(subset=['Geocoded Address'])
df.head()
```

```python

df['longitude'] = df.apply(lambda row: float(row['Geocoded Address'].split()[1][1:]), axis=1)
df['latitude'] = df.apply(lambda row: float(row['Geocoded Address'].split()[2][:-1]), axis=1)
```

```python
df = df[['State Code', 'longitude', 'latitude', 'Courses Available']]
```

```python
df.to_csv('../data/medical_facilities.csv')
```
